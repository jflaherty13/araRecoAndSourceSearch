```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Run Number',fontsize=13)
ax1.set_ylabel('exp', color=color)
ax1.plot(CW["run"],CW["phase_var"],label="Phase variance",linewidth=1,color=color)
ax1.plot(CW["run"],CW["baseline"],"--",label="Peaks over baselines",linewidth=1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('#calpulsers/#events', color=color,fontsize=13)  # we already handled the x-label with ax1
ax2.plot(calpul["Run Number"],0*calpul["Ratio"],label="Peaks over baselines",linewidth=1,color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.xlim(1450,1910)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
```
